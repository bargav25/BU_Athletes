```python
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
import copy
from sklearn.metrics import silhouette_score
import seaborn as sns
import matplotlib.pyplot as plt
from statistics import median

df_copy = copy.deepcopy(df)
df_copy = df_copy.dropna(subset = ['SATCFORAI', 'HSGPA', 'Last Cumulative BU GPA']) 
df_pd = df_copy[['SATCFORAI', 'HSGPA', 'Last Cumulative BU GPA']].values

K = range(5, 10)
fits = []
score = []

for k in K:
    model = KMeans(n_clusters = k).fit(df_pd)
    fits.append(model)
    score.append(silhouette_score(df_pd, model.labels_))

maximum = score.index(max(score))
labels = fits[maximum].labels_

x = []
y = []
z = []

for row in df_pd:
    x.append(row[0])
    y.append(row[1])
    z.append(row[2])
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.set_title("3D Scatter Plot with Labeled Clusters of Features: HSGPA, College GPA, HS SAT")
ax.scatter(x, y, z, c=labels)

track = {i:[] for i in range(min(labels), max(labels) + 1)}
for x in range(len(labels)):
    track[labels[x]].append(df_pd[x])

for key in track.keys():
    sat = []
    hs_gpa = []
    c_gpa = []

    for point in track[key]:
        sat.append(point[0])
        hs_gpa.append(point[1])
        c_gpa.append(point[2])
    print("Cluster", (key + 1), "stats:")
    print("SAT Average:", round((sum(sat)/len(sat))), "Median:", str((median(sat))))
    print("HS GPA Average:", round((sum(hs_gpa)/len(hs_gpa)), 3), "Median:", str((median(hs_gpa))))
    print("C GPA Average:", round((sum(c_gpa)/len(c_gpa)), 3), "Median:", str(median(c_gpa)))
    print()
    print()

```

```python
# Load libraries
import pandas as pd
import copy
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
#split dataset in features and target variable
temp = copy.copy(df)
feature_cols = ['Language', 'Team Gender', 'HSGPA', 'SATCFORAI']
temp['Team Gender'] = temp['Team Gender'].map({'F': 1, 'M': 0})
temp['Language'] = temp['Language'].map({'English': 1, 'Non-English': 0})
temp = temp.dropna(subset = ['SATCFORAI', 'HSGPA', 'Language', 'Team Gender']) 
labels = []
for row in temp['Last Cumulative BU GPA']:
    if row >= 3.3:
        labels.append(1)
    elif row >= 2.3:
        labels.append(2)
    elif row >= 1.3:
        labels.append(3) 
    else:
        labels.append(4)
#print(labels)
    

X = temp[feature_cols] # Features

y = labels # Target variable
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1) # 70% training and 30% test
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
